# Ques: 17- To obtain the feature weights from the linear model and find the features having the lowest and highest weights

```python
#comparing the Effect of Regularization

def weights_def(model, feature, col_name):
    #this function returns the weight of every feature
  weights = pd.Series(model.coef_, feature.columns).sort_values()
  weights_df = pd.DataFrame(weights).reset_index()
  weights_df.columns = ['Features', col_name]
  weights_df[col_name].round(3)
  return weights_df
linear_model_weights = weights_def(lr2, X_train, 'Linear_Model_Weight')
ridge_weights_df = weights_def(rid, X_train, 'Ridge_Weight')
lasso_weights_df = weights_def(las, X_train, 'Lasso_weight')
```

```python
#Combining Weights on the basis of Features
final_weights = pd.merge(linear_model_weights, ridge_weights_df, on='Features')
final_weights = pd.merge(final_weights, lasso_weights_df, on='Features')
```

```python
final_weights
```

RH_2 has the lowest weight = -0.479066

RH_1 has the highest weight=  0.545721
